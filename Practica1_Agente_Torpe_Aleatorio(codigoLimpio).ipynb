## Agente Torpe (acciones aleatorias)
El presente código muestra la manera "limpia" de codificar que se seguirá en las próximas prácticas, así como un código final que contará el número de veces que el agente ha tenido exito con estas acciones aleatorias

#### Librerías

```python
import numpy as np
import matplotlib.pyplot as plt
import gymnasium as gym
from gymnasium.envs.toy_text.frozen_lake import generate_random_map
```

#### Crear una clase para el agente

```python
class AgenteTorpe:
    def __init__(self):
        self.ambienteInterno = gym.make('FrozenLake-v1', is_slippery=False, render_mode='rgb_array', map_name="4x4")
        
    def select_action(self):
        return self.ambienteInterno.action_space.sample()
```

#### Función para ejecutar un conjunto de acciones

```python
def test(agente, ambiente):
    ambiente.reset()
    is_done = False
    t = 0
    
    while not is_done:
        action = agente.select_action()
        estado, recompensa, is_done, trunc, info = ambiente.step(action)
        t += 1
    
    return (estado, recompensa, t)
```

#### Ejecutar un conjunto de acciones hasta su fin N episodios y contar el número de éxitos

```python
episodios = 1000
resueltos = 0

# Crear el agente y el ambiente
agente = AgenteTorpe()
env = gym.make('FrozenLake-v1', is_slippery=False, render_mode='rgb_array', map_name="4x4")

for episodio in range(episodios):
    estado, recompensa, t = test(agente, env)
    
    if estado==15:
        resueltos += 1
        
print("El agente resolvió el ambiente en {} de {} episodios".format(resueltos, episodios))
```

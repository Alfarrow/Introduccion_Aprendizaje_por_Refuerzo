Librerías a usar

```python
import gymnasium as gym
from gymnasium.envs.toy_text.frozen_lake import generate_random_map
```

Especificar el entorno que se usará que será "Frozen Lake"

```python
# Configurar un mapa personalizado
mapa = ["SFFF", "FHFH", "FFFH", "HFFG"]
# Crear ambiente
env = gym.make('FrozenLake-v1', is_slippery=False, render_mode='rgb_array', desc=mapa)
env.reset()
```

Se muestra el estado inicial

```python
# Mostrar imagen con matplotlib
import matplotlib.pyplot as plt

array = env.render()
plt.imshow(array)
```

Se muestran el número de acciones posibles y su tipo

```python
print(env.action_space)
```

Se muestra la observación del agente en este momento

```python
print(env.observation_space)
```

```python
print(env.P)
```

Correr una secuencia de steps hasta que el juego acabe o se alcance la meta

```python
is_done = False
t = 0

while not is_done:
    action = env.action_space.sample()
    state, reward, is_done, truncado, info = env.step(action)
    env.render()
    t += 1
    print(env.desc)
    
```

Mostrar Resultado

```python
print('El último estado fue: ', state)
print('La recompensa fue: ', reward)
print('Número de time-steps: ', t)
```

Mostrar Estado Final

```python
plt.imshow(env.render())
```
